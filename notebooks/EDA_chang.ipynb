```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ap=pd.read_csv('chang_amazonprime_34.csv')
yt=pd.read_csv('chang_youtube_5.csv')
nf=pd.read_csv('jeq004_netflix_1080p_1x_vpn_mac_clean_20210104-21.csv')
```

```python
def plot(data):
    fig, ax = plt.subplots()
    ax.plot(data['Time'],data['2->1Bytes'])
    plt.figtext(0.66,0.89, data.describe().loc[['mean', '50%', 'std'],'2->1Bytes'].to_string())
    plt.ylabel('2->1Bytes')
    plt.xlabel('time')
    plt.title('Time vs 2->1Bytes')
```

```python
plot(ap)
```

```python
plot(yt)
```

```python
plot(nf)
```
